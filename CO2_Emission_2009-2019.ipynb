```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import datetime as dt
```

```python
# Name of the CSV file
file = 'Carbon_data.xlsx'
```

```python
df = pd.read_excel(file)
```

```python
df.head()
```

## Cleandata change year month date to year month


```python
df["Month"]=pd.to_datetime(df["Month"]).dt.strftime('%Y-%m')
```

```python
df=df.set_index('Month')
```

```python
df=df.T
```

```python
# Plot the CO2 Emissions

plt.figure(figsize=(20,10))

plt.plot (df.columns, df.loc["Coal, Including Coal Coke Net Imports, CO2 Emissions",:],color="black", label="Coal CO2 Emission")

plt.plot (df.columns, df.loc["Petroleum, Excluding Biofuels, CO2 Emissions",:],color="brown", label="Petroleum CO2 Emission")

plt.plot (df.columns, df.loc["Total Energy CO2 Emissions",:],color="green", label="Total Energy CO2 Emission")

plt.plot (df.columns, df.loc["Natural Gas, Excluding Supplemental Gaseous Fuels, CO2 Emissions",:], color="red", label="Natural Gas CO2 Emission")

plt.title("Energy CO2 emission")
plt.xlabel("Year")
plt.xticks(rotation=90)
plt.ylabel("Million metric tons of Carbon Dioxid")

plt.legend()
plt.savefig('Energy Carbon Dioxide.png')
plt.show()
```

```python
# Plot the Coal CO2 Emissions
plt.figure(figsize=(20,10))

plt.plot (df.columns, df.loc["Coal, Including Coal Coke Net Imports, CO2 Emissions",:].rolling(window=12).mean(), color="black", label="Coal CO2 Emission")

plt.plot (df.columns, df.loc["Petroleum, Excluding Biofuels, CO2 Emissions",:].rolling(window=12).mean(),color="brown", label="Petroleum CO2 Emission")

plt.plot (df.columns, df.loc["Total Energy CO2 Emissions",:].rolling(window=12).mean(),color="green", label="Total Energy CO2 Emission")

plt.plot (df.columns, df.loc["Natural Gas, Excluding Supplemental Gaseous Fuels, CO2 Emissions",:].rolling(window=12).mean(), color="red", label="Natural Gas CO2 Emission")
plt.title("Energy CO2 emission")
plt.xlabel("Year")
plt.xticks(rotation=90)
plt.ylabel("Million metric tons of Carbon Dioxid")

plt.legend()
plt.savefig('Energy Carbon Dioxide RM-12.png')
plt.show()
```
